```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
ReportType = 'Daily'
Format = 'csv'
StartDate = '9/04/2010'
EndDate = '9/06/2010'
Country = 'can'
province = ['NB', 'NS', 'NL', 'PE']
# province = ['QC', 'ON']
# province = ['NB', 'NS', 'NL', 'PE', 'ON', 'QC']
# province = ['BC', 'YK']
```

```python
hurr_url = f'https://data.cocorahs.org/export/exportreports.aspx?ReportType={ReportType}&Format={Format}&ReportDateType=reportdate&StartDate={StartDate}&EndDate={EndDate}&country={Country}&province={",".join(province)}'
```

```python
hurr = pd.read_csv(hurr_url)
```

```python
hurr.drop(columns=['NewSnowDepth','NewSnowSWE','TotalSnowDepth','TotalSnowSWE'], axis=1, inplace= True)
hurr['ReportType'] = ReportType
hurr['Hurricane_Name'] = 'EARL'
```

```python
cnx = sql.connect("Canadian_Hurricane_Impact.db") #input Database Name
```

```python
hurr.to_sql('Complete_Impact', cnx, if_exists='append', index = False) # Exports data to an SQL Lite Database
```

```python
cnx.close()
```

# Adjust Precip amounts to Metric

```python
cnx = sql.connect("Canadian_Hurricane_Impact.db") #input Database Name
hurr_db = pd.read_sql_query('SELECT * FROM Complete_Impact ', cnx)
cnx.close()
```

```python
hurr_db['TotalPrecipAmt'] = pd.to_numeric(hurr_db['TotalPrecipAmt'], errors='coerce').fillna(0, downcast='float')
hurr_db['TotalPrecipAmt'] = hurr_db['TotalPrecipAmt'] * 25.4
```

```python
cnx = sql.connect("Canadian_Hurricane_Impact.db") #input Database Name
hurr_db.to_sql('Complete_Impact', cnx, if_exists='replace', index = False) # Exports data to an SQL Lite Database
cnx.close()
```
